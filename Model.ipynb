PCA mixes all numerical features into new components.

putting transaction amount and time features inside PCA, they will get combined with others.

This can hide their direct interpretability (e.g., “high transaction amount = risky” won’t be directly visible anymore)

```python
#Split numerical columns
num_pca_features = ['lat','long','merch_lat','merch_long','city_pop']
num_direct_features = ['amt','hour','day','month','age']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.impute import SimpleImputer
import category_encoders as ce
```

```python
##Piplines

#numerical piplines
num_pca_pipeline=Pipeline(
    steps=[
        ("scaler",StandardScaler()),
        ("imputer",SimpleImputer(strategy='median')),
        ("pca",PCA(n_components=0.95))  
    ]
)

num_direct_pipeline=Pipeline(
    steps=[
        ("Scaler",StandardScaler()),
        ("imputer",SimpleImputer(strategy='median'))
    ]
)



##categorical piplines
# High-cardinality categorical pipeline → Target Encoding
cat_high_card_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("target_enc", ce.TargetEncoder())
])

# Low-cardinality categorical pipeline → OneHot
cat_low_card_pipeline = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])



preprocessor=ColumnTransformer([
    ('num_pca_pipeline', num_pca_pipeline,num_pca_features),
    ('num_direct_pipline',num_direct_pipeline,num_direct_features),
    ('cat_high_card_pipeline',cat_high_card_pipeline,cat_high_card),
    ('cat_low_card_pipline',cat_low_card_pipeline,cat_low_card)
]
)
```

```python
x=df.drop(labels=['is_fraud'], axis=1)
y=df['is_fraud']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
# Fit and transform just to check
x_transformed = preprocessor.fit_transform(x_train,y_train)

# Convert to dense numpy array

print("Original shape:", x_train.shape)
print("Transformed shape:", x_transformed.shape)
```

```python
from imblearn.pipeline import Pipeline
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# Example: Logistic Regression pipeline
log_reg_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),     # your column transformer
    ("smote", SMOTE(random_state=42)),  # handle imbalance
    ("model", LogisticRegression(max_iter=1000))
])


# Example: Random Forest pipeline
rf_pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("smote", SMOTE(random_state=42)),
    ("model", RandomForestClassifier(n_estimators=100, random_state=42))
])
```

```python
rf_pipeline.fit(x_train, y_train)
y_pred = rf_pipeline.predict(x_test)
```

```python
from collections import Counter

print("Before SMOTE:", Counter(y_train))   # original imbalance
X_res, y_res = SMOTE(random_state=42).fit_resample(preprocessor.fit_transform(x_train, y_train), y_train)
print("After SMOTE:", Counter(y_res))      # balanced dataset
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Predictions
y_pred = rf_pipeline.predict(x_test)

# Metrics
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
