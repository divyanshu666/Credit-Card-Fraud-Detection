```python
#INDEPENDENT AND DEPENDENT FEATURES
x=df.drop('Class', axis=1)
y=df['Class']
```

```python
## transform the dataset
from imblearn.over_sampling import SMOTE
oversample=SMOTE()
x, y=oversample.fit_resample(x,y)
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df[['Time', 'Amount']] = scaler.fit_transform(df[['Time', 'Amount']])
```

```python
import pandas as pd
df_x = pd.DataFrame(x, columns=x.columns)   # use column names properly
df_y = pd.DataFrame(y, columns=["Class"]) 
oversample_df=pd.concat([df_x,df_y], axis=1)

oversample_df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(oversample_df['V1'],oversample_df['V2'], c=oversample_df['Class'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
x_train.shape, y_train.shape
```

```python
import seaborn as sns
sns.heatmap(x_train.corr())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
```

```python
logreg=LogisticRegression(max_iter=1000)
logreg.fit(x_train,y_train)
y_pred=logreg.predict(x_test)

precision=precision_score(y_test, y_pred)
roc_auc=roc_auc_score(y_test,y_pred)
print('precision',precision)
print('roc_auc',roc_auc)
```
